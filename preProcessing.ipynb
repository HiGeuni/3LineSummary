```python
# imports modules
import json
import random
import pickle

file_path = "./data.json"
random.seed(42)
```

```python
with open("train_original.json") as f:
    data = json.load(f)
```

```python
document = data["documents"]
doc = []
num_avg_word = 0

exception_word = ["@", "일보", "기자", "사진", "자료=", "네이버에서", "바로가기", "문의전화", "최저금리", "http"]
```

```python
# List Doc : include all sentence in json
for i in range(len(document)):
    for j in range(len(document[i]["text"])):
        if len(document[i]["text"][j]) == 0:
            continue
        content = document[i]["text"][j][0]["sentence"]    
        flag = False
        for word in exception_word:
            if word in content:
                flag = True
                break
        if not flag:
            doc.append(content)
```

```python
# Calculate Word Num
def getWordNum(doc):
    num_word = 0
    for i in doc:
        num_word += len(i.split(" "))
    return num_word
```

```python
# Calculate Word Num json
def getWordNumJson(doc):
    num_word = 0
    for i in range(len(doc)):
        for j in doc[i]["content"]:
            num_word += len(j.split(" "))
    return num_word
```

```python
# Calculate Average Number of Word per Document
num_word = getWordNum(doc)
print( num_word / len(document))
```

```python
# 자원의 제약이 있으므로, 랜덤하게 문서를 230개를 뽑아서 데이터셋으로 설정하였습니다.
# Json으로 만들기 위해서 Dictionary 를 사용했습니다.
my_list = list(range(0, len(document)))
random.shuffle(my_list)

sentenceList = []

for i in range(230):
    dc = {}
    dc["title"] = document[my_list[i]]["title"]
    sl = []
    for j in range(2, len(document[my_list[i]]["text"])):
        if len(document[my_list[i]]["text"][j]) == 0:
            continue
        content = document[my_list[i]]["text"][j][0]["sentence"]    
        flag = False
        for word in exception_word:
            if word in content:
                flag = True
                break
        if not flag:
            sl.append(content)
    dc["content"] = sl
    sentenceList.append(dc)
```

```python
getWordNumJson(sentenceList)
```
